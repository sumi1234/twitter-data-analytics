```python
import pandas as pd
import numpy as np

data_frame = pd.read_csv("C:\Users\Sumedha\Documents\world_cup_final.csv")

print data_frame.head(5)
```

```python
import nltk

from nltk.corpus import stopwords

from nltk import FreqDist

stop = stopwords.words('english')
text = data_frame['tweet_text']
tokens = []

for txt in text.values:
    tokens.extend([t.lower().strip(":,.") for t in txt.split()])

filtered_tokens = [w for w in tokens if not w in stop]
```

```python
freq_dist = nltk.FreqDist(filtered_tokens)

freq_dist
```

```python
from collections import Counter

#user_col = data_frame.pop('user_name').apply(pd.Series)

authors = data_frame['user_name']

authors_counter = Counter(authors.values)

print "Most frequent authors"
print "\n".join(["{0}\t{1}".format(a,f) for a,f in authors_counter.most_common(25)])
print
```

```python
print "After removing stopwords"
#we could do further preprocessing by stemming etc

for t in nltk.corpus.stopwords.words('english'):
    if t in tokens_counter:
        tokens_counter.pop(t)
    
tokens_counter.most_common(100)
```
